```python
import csv
from tabulate import tabulate

# Function to read tables from CSV files
def read_tables():
    tables = []
    file_names = ["grades.csv", "class_students.csv", "rabbytes_club_students.csv", "rabbytes_data.csv"]
    for file_name in file_names:
        with open(file_name, "r") as csv_file:
            reader = csv.reader(csv_file)
            table = [row for row in reader]
            tables.append(table)
    return tables

# Function to list tables
def list_tables(tables):
    print("\n  Index    Columns    Rows")
    print("-------  ---------  ------")
    for index, table in enumerate(tables):
        num_columns = len(table[0]) if table else 0
        num_rows = len(table)
        print(f"{index:<7}  {num_columns:<9}  {num_rows:<6}")

# Function to display a table given its index
def display_table(tables):
    while True:
        try:
            index = int(input("Choose a table index (to display): "))
            if 0 <= index < len(tables):
                print(tabulate(tables[index], headers="firstrow", tablefmt="grid"))
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a valid table index.")

# Function to duplicate a table given its index
def duplicate_table(tables):
    while True:
        try:
            index = int(input("Choose a table index (to duplicate): "))
            if 0 <= index < len(tables):
                tables.append([row[:] for row in tables[index]])
                print(f"Table {index} duplicated successfully.")
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a valid table index.")

# Main function to run the menu
def main():
    tables = read_tables()
    while True:
        print("\n==================================")
        print("Enter your choice:")
        print("1. List tables.")
        print("2. Display table.")
        print("3. Duplicate table.")
        print("0. Quit.")
        print("==================================")
        choice = input("Enter your choice: ")

        if choice == "1":
            list_tables(tables)
        elif choice == "2":
            display_table(tables)
        elif choice == "3":
            duplicate_table(tables)
        elif choice == "0":
            print("Goodbye!")
            break
        else:
            print("Invalid choice. Please enter a valid option.")

if __name__ == "__main__":
    main()
```
