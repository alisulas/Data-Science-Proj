```python
# t2_table_menu_create_delete.py

import pandas as pd
from tabulate import tabulate

# List of tables with their file paths
table_files = [
    "supplementary_files/grades.csv",
    "supplementary_files/class_students.csv",
    "supplementary_files/rabbytes_club_students.csv",
    "supplementary_files/rabbytes_data.csv"
]

# Load tables into a dictionary for easy access
tables = {index: pd.read_csv(file) for index, file in enumerate(table_files)}

def list_tables():
    # Print the list of tables with index, columns, and rows
    table_summary = [{"Index": i, "Columns": len(df.columns), "Rows": len(df)} for i, df in tables.items()]
    print(tabulate(table_summary, headers="keys"))



def display_table():
    try:
        index = int(input("Choose a table index (to display): "))
        if index in tables:
            print(tabulate(tables[index], headers="keys", tablefmt="grid"))
        else:
            print("Incorrect table index. Try again.")
            display_table()  # Retry if the index is invalid
    except ValueError:
        print("Invalid input. Please enter a number.")
        display_table()  # Retry if the input is not a valid number






def duplicate_table():
    try:
        index = int(input("Enter table index to duplicate: "))
        if index in tables:
            new_index = len(tables)
            tables[new_index] = tables[index].copy()
            print(f"Table at index {index} duplicated at index {new_index}.")
        else:
            print("Invalid index. Please try again.")
    except ValueError:
        print("Invalid input. Please enter a number.")



def create_table():
    # Step 1: Ask for a valid table index to create from
    while True:
        try:
            index = int(input("Choose a table index (to create from): "))
            if index in tables:
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")
    
    # Step 2: Display the columns of the selected table
    selected_table = tables[index]
    num_columns = len(selected_table.columns)
    print(f"Columns in table {index}:")
    for i, col in enumerate(selected_table.columns):
        print(f"{i}: {col}")
    
    # Step 3: Ask for the indices of the columns to keep
    while True:
        try:
            col_indices = input("Enter the comma-separated indices of the columns to keep: ").split(',')
            col_indices = [int(i.strip()) for i in col_indices]
            
            # Check if all indices are valid
            if all(0 <= i < num_columns for i in col_indices):
                # Step 4: Create the new table with selected columns
                new_table = selected_table.iloc[:, col_indices]
                new_index = max(tables.keys()) + 1
                tables[new_index] = new_table
                print(f"New table created at index {new_index}.")
                break
            else:
                print("Some indices are out of range. Try again.")
        except ValueError:
            print("Invalid input. Please enter comma-separated numbers only.")

def delete_table():
    # Step 1: Ask for a valid table index to delete
    while True:
        try:
            index = int(input("Choose a table index (for table deletion): "))
            if index in tables:
                # Step 2: Confirm deletion with the user
                confirm = input(f"Are you sure you want to delete table {index}? (yes/no): ").strip().lower()
                if confirm == 'yes':
                    del tables[index]  # Delete the table from the dictionary
                    print(f"Table {index} has been deleted.")
                else:
                    print("Deletion canceled.")
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")



def main_menu():
    while True:
        print("="*34)
        print("Enter your choice:")
        print("1. List tables.")
        print("2. Display table.")
        print("3. Duplicate table.")
        print("4. Create table.")
        print("5. Delete table.")
        print("0. Quit.")
        print("="*34)

        choice = input().strip()
        
        if choice == '1':
            list_tables()
        elif choice == '2':
            display_table()
        elif choice == '3':
            duplicate_table()
        elif choice == '4':
            create_table()
        elif choice == '5':
            delete_table()  # This function will be implemented for deleting tables
        elif choice == '0':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter 1, 2, 3, or 0.")
            


if __name__ == "__main__":
    main_menu()
```

```python
# t3_table_menu_column_restore.py

import pandas as pd
from tabulate import tabulate

# List of tables with their file paths
table_files = [
    "supplementary_files/grades.csv",
    "supplementary_files/class_students.csv",
    "supplementary_files/rabbytes_club_students.csv",
    "supplementary_files/rabbytes_data.csv"
]

# Load tables into a dictionary for easy access
tables = {index: pd.read_csv(file) for index, file in enumerate(table_files)}

def list_tables():
    # Print the list of tables with index, columns, and rows
    table_summary = [{"Index": i, "Columns": len(df.columns), "Rows": len(df)} for i, df in tables.items()]
    print(tabulate(table_summary, headers="keys"))



def display_table():
    try:
        index = int(input("Choose a table index (to display): "))
        if index in tables:
            print(tabulate(tables[index], headers="keys", tablefmt="grid"))
        else:
            print("Incorrect table index. Try again.")
            display_table()  # Retry if the index is invalid
    except ValueError:
        print("Invalid input. Please enter a number.")
        display_table()  # Retry if the input is not a valid number






def duplicate_table():
    try:
        index = int(input("Enter table index to duplicate: "))
        if index in tables:
            new_index = len(tables)
            tables[new_index] = tables[index].copy()
            print(f"Table at index {index} duplicated at index {new_index}.")
        else:
            print("Invalid index. Please try again.")
    except ValueError:
        print("Invalid input. Please enter a number.")



def create_table():
    # Step 1: Ask for a valid table index to create from
    while True:
        try:
            index = int(input("Choose a table index (to create from): "))
            if index in tables:
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")
    
    # Step 2: Display the columns of the selected table
    selected_table = tables[index]
    num_columns = len(selected_table.columns)
    print(f"Columns in table {index}:")
    for i, col in enumerate(selected_table.columns):
        print(f"{i}: {col}")
    
    # Step 3: Ask for the indices of the columns to keep
    while True:
        try:
            col_indices = input("Enter the comma-separated indices of the columns to keep: ").split(',')
            col_indices = [int(i.strip()) for i in col_indices]
            
            # Check if all indices are valid
            if all(0 <= i < num_columns for i in col_indices):
                # Step 4: Create the new table with selected columns
                new_table = selected_table.iloc[:, col_indices]
                new_index = max(tables.keys()) + 1
                tables[new_index] = new_table
                print(f"New table created at index {new_index}.")
                break
            else:
                print("Some indices are out of range. Try again.")
        except ValueError:
            print("Invalid input. Please enter comma-separated numbers only.")

def delete_table():
    # Step 1: Ask for a valid table index to delete
    while True:
        try:
            index = int(input("Choose a table index (for table deletion): "))
            if index in tables:
                # Step 2: Confirm deletion with the user
                confirm = input(f"Are you sure you want to delete table {index}? (yes/no): ").strip().lower()
                if confirm == 'yes':
                    del tables[index]  # Delete the table from the dictionary
                    print(f"Table {index} has been deleted.")
                else:
                    print("Deletion canceled.")
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def delete_column():
    # Step 1: Ask for a valid table index to delete a column from
    while True:
        try:
            index = int(input("Choose a table index (for column deletion): "))
            if index in tables:
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")
    
    # Step 2: Display the columns of the selected table
    selected_table = tables[index]
    num_columns = len(selected_table.columns)
    print(f"Columns in table {index}:")
    for i, col in enumerate(selected_table.columns):
        print(f"{i}: {col}")
    
    # Step 3: Ask for the column index to delete
    while True:
        try:
            col_index = int(input("Enter the index of the column to delete: "))
            if 0 <= col_index < num_columns:
                # Get the column name by index
                column_name = selected_table.columns[col_index]
                # Delete the column
                tables[index] = selected_table.drop(columns=[column_name])
                print(f"Column '{column_name}' deleted from table {index}.")
                break
            else:
                print("Column index out of range. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")





# Dictionary to store deleted tables for restoration
deleted_tables = {}

def delete_table():
    # Existing delete_table function with modifications to store deleted tables
    while True:
        try:
            index = int(input("Choose a table index (for table deletion): "))
            if index in tables:
                confirm = input(f"Are you sure you want to delete table {index}? (yes/no): ").strip().lower()
                if confirm == 'yes':
                    # Move the table to deleted_tables before deletion
                    deleted_tables[index] = tables.pop(index)
                    print(f"Table {index} has been deleted and is available for restoration.")
                else:
                    print("Deletion canceled.")
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def restore_table():
    # Step 1: Check if any tables have been deleted
    if not deleted_tables:
        print("No tables are currently available for restoration.")
        return
    
    # Step 2: Ask for a valid table index to restore
    while True:
        try:
            index = int(input("Choose a table index (for restoration): "))
            if index in deleted_tables:
                # Restore the table from deleted_tables to tables
                tables[index] = deleted_tables.pop(index)
                print(f"Table {index} has been restored.")
                break
            else:
                print("Incorrect table index. Try again.")
        except ValueError:
            print("Invalid input. Please enter a number.")

def main_menu():
    while True:
        print("="*34)
        print("Enter your choice:")
        print("1. List tables.")
        print("2. Display table.")
        print("3. Duplicate table.")
        print("4. Create table.")
        print("5. Delete table.")
        print("6. ta from table.")
        print("7. Restore table.")  # New option for restoring a table
        print("0. Quit.")
        print("="*34)

        choice = input().strip()
        
        if choice == '1':
            list_tables()
        elif choice == '2':
            display_table()
        elif choice == '3':
            duplicate_table()
        elif choice == '4':
            create_table()
        elif choice == '5':
            delete_table()
        elif choice == '6':
            delete_column()
        elif choice == '7':
            restore_table()  # Calls the restore_table function
        elif choice == '0':
            print("Exiting program.")
            break
        else:
            print("Invalid choice. Please enter a number from the menu.")

            


if __name__ == "__main__":
    main_menu()
```
